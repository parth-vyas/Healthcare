```python
import pandas as pd
```

```python
df = pd.read_csv("liver_patient.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(how='any',inplace=True)
```

```python
print(df["Gender"].unique())
print(df["Dataset"].unique())
```

```python
df["Gender"] = df["Gender"].replace(to_replace={'Male':1,'Female':0})
df["Dataset"] = df["Dataset"].replace(to_replace={1:0,2:1})
```

```python
X = df.iloc[:,:10]
y = df.iloc[:,-1]
```

```python
print(y.unique())
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=30)
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
#hyperparameter tuning
n_estimators = [int(i) for i in np.linspace(100,1000,10)]
criterion = ['gini','entropy']
max_depth = [int(i) for i in np.linspace(100,1000,10)]
min_samples_split = [2,3,5,6,8,9]
min_samples_leaf = [1,2,8,6,5,7]
max_features = ['auto','sqrt','log2']

param_grid = {'n_estimators':n_estimators,
              'criterion':criterion,
              'max_depth':max_depth,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf,
              'max_features':max_features
}
print(param_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_cv = RandomizedSearchCV(estimator=rf,param_distributions=param_grid,n_iter=100,cv=3,verbose=2,n_jobs=1)
rf_cv.fit(X_train,y_train)
```

```python
best_param = rf_cv.best_estimator_
```

```python
y_pred = best_param.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
print("Accuracy score:{}".format(accuracy_score(y_test,y_pred)))
```

```python
import pickle
file = open("random_forest_liver_model.pkl","wb")
pickle.dump(rf_cv,file)
```
